```python
# Feature engineering for ADHD, Anxiety, Self-esteem, and Depression scores
ADHD = ['Use_Of_Social_Media_Without_Purpose', 'Distraction_Level']
smmh['adhd_score'] = smmh[ADHD].sum(axis=1)

Anxiety = ['Restlessness_Level_Without_Social_Media', 'Worry_Level']
smmh['anxiety_score'] = smmh[Anxiety].sum(axis=1)

Self_esteem = ['Comparison_To_Successful_People_Level']
smmh['self_esteem_score'] = smmh[Self_esteem].sum(axis=1)

Depression = ['Interest_Fluctuation_Level', 'Sleep_Issues_Level', 'Appetite_Issues_Level']
smmh['depression_score'] = smmh[Depression].sum(axis=1)

#Selecting relevant features and target columns
features = ADHD + Anxiety + Self_esteem + Depression
targets = ['adhd_score', 'anxiety_score', 'self_esteem_score', 'depression_score']
```

```python
#one score containing all 4 target columns
total = ['adhd_score', 'anxiety_score', 'self_esteem_score', 'depression_score']

smmh['mental_health_prob_score'] = smmh[total].sum(axis =1)
```

```python
# Function to calculate mental health scores based on social media platforms and usage
def calculate_scores(social_media_platforms, daily_social_media_usage_time):

    #Initialize scores
    adhd_score = 0
    self_esteem_score = 0
    anxiety_score = 0
    depression_score = 0

    #Check if user has used more than 4 social media platforms
    if len(social_media_platforms) > 4:  # e.g. if they use more than 4 platforms
        adhd_score += 1  #Increase ADHD score
        anxiety_score += 1  #Increase Anxiety score
        depression_score += 1  #Increase Depression score

    #Check for daily social media usage time
    if daily_social_media_usage_time in ['4 to 6 hours']:
        anxiety_score += 1  #Increase Anxiety score for higher usage time

    #Total score calculation
    total_score = adhd_score + self_esteem_score + anxiety_score + depression_score

    #Return calculated scores
    return adhd_score, self_esteem_score, anxiety_score, depression_score, total_score
```

```python
#Function to update mental_health_prob_score based on conditions
def update_mental_health_score(row):
    #Split platforms and check how many are used
    platforms = row['Social_Media_Platforms'].split(';')
    daily_time = row['Daily_Social_Media_Time']

    #Initialize score adjustments
    adhd_score = row['adhd_score']
    anxiety_score = row['anxiety_score']
    depression_score = row['depression_score']

    #Check if the user is using more than 4 platforms
    if len(platforms) > 4:
        adhd_score += 1
        anxiety_score += 1
        depression_score += 1

    #Check if daily usage is '4 to 6 hours'
    if daily_time == '4 to 6 hours':
        anxiety_score += 1
        depression_score += 1

    #Update mental health probability score (sum of all scores)
    total_score = adhd_score + row['self_esteem_score'] + anxiety_score + depression_score
    return total_score

#Apply the function to each row and update the mental_health_prob_score
smmh['mental_health_prob_score'] = smmh.apply(update_mental_health_score, axis=1)

#Display the updated DataFrame
smmh[['Social_Media_Platforms', 'Daily_Social_Media_Time', 'mental_health_prob_score']].head()
```

Identifying clusters of respondents with similar social media usage and mental health profiles using clustering algorithm


```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

inertia = [] #Initialize a list to store inertia values
k_range = range(1, 10) #Define the range of clusters to test

for k in k_range: # Iterate over the range of cluster values
    kmeans = KMeans(n_clusters=k, random_state=42) # Initialize KMeans with current cluster count
    kmeans.fit(smmh) #Fit KMeans model to the data
    inertia.append(kmeans.inertia_) #Append the inertia (sum of squared distances) to the list

plt.plot(k_range, inertia, marker='o')  # Plot inertia values against the number of clusters
plt.xlabel('Number of Clusters (3)')
plt.ylabel('Inertia')
plt.title('Elbow Method for Optimal 3')
plt.show()
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#standardizing data
data_for_clustering = smmh[['mental_health_prob_score']]
data_standardized = StandardScaler().fit_transform(data_for_clustering)

#KMeans clustering
cluster_labels = KMeans(n_clusters=3, random_state=0).fit_predict(data_standardized)
smmh['Cluster'] = cluster_labels

#creating a random y-axis for visualization purposes
y_random = np.random.randn(len(data_standardized))

#plotting the standardized mental health score with cluster labels
plt.scatter(data_standardized, y_random, c=cluster_labels, cmap='rainbow')
plt.title('Clusters of Respondents')
plt.xlabel('Standardized Mental Health Score')
plt.show()
```

**Advanced Statistics**

```python
#creating new var based on mental_health_prob_score.

# For cut off, 33. This variable is based on 10 other variables in which 8 variables ranges from 1-5 and 2 other vars such as social media platforms and daily social media time.
# For social media platforms, if the user has used more than 4 social media platforms, then adhd_score, anxiety_score and depression_score , each increases by 1 and adds up in mental_health_prob_score(i.e +3).
# For daily social media time, if the user has used social media from 4 to 6 hours(5)in new dataset, then anxiety_score and depression_score , each increases by 1 and adds up in mental_health_prob_score(i.e. +2).
#If a person answers all of the questions as neutral, it would be 24.
# If a person answers all of them as 4, it would be 32. I choose the number in between, which is 28.
#Then I add 28 with 3 and 2, I get 33 as total.
# cutoff as 33. If it is 33 and higher, that means this person needs psychological counselling.


# Function to map the score based on cut-off value (33)
def map_score(i):
  if i < 33: #if the score is less than 33, no counselling is needed
    return "0"
  elif i >= 33: #if score is 33 or higher, counselling is needed
    return "1"

#Applying the function to create 'needs_counselling' column
clusters_smmh['needs_counselling']= clusters_smmh['mental_health_prob_score'].apply(lambda i: map_score(i))

#Converting the 'needs_counselling' column to integer type
clusters_smmh['needs_counselling']= clusters_smmh['needs_counselling'].astype('int64')
```

```python
#creating new initial dataframe to run the analysis.
needs_counselling = clusters_smmh[['Daily_Social_Media_Time', 'Age', 'Gender', 'adhd_score', 'anxiety_score', 'self_esteem_score', 'depression_score', 'mental_health_prob_score', 'needs_counselling']]
```

**LOGISTIC REGRESSION MODEL**

```python
#applying Logistic Regression Model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

#taking ip and op(features and target vars)
x = needs_counselling[['Daily_Social_Media_Time', 'Age', 'Gender', 'adhd_score', 'anxiety_score', 'self_esteem_score', 'depression_score']]
y = needs_counselling['needs_counselling']

#splitting data into train and test set
x_train, x_test, y_train, y_test = train_test_split(x, y , test_size=0.2,random_state=22)
```

```python
#running regression analysis
model = LogisticRegression()
model.fit(x_train, y_train)

predicted= model.predict(x_test)
accuracy= model.score(x_test,y_test)

#confusion matrix.
from sklearn import metrics
cm = metrics.confusion_matrix(y_test, predicted)
print(cm)
```

The confusion matrix indicates that the model correctly identified 46 true positives, made 1 false positive, made 1 false negative, and correctly identified 1 true negative.
